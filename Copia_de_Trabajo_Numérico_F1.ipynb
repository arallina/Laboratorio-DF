$\frac{d\vec{p_1}}{dt}=F_g\hat r \neq 0$ el momento lineal no se conserva debido a la fuerza de gravedad.

Momento Angular

$\frac{d\vec{L_1^o}}{dt}=r\hat {r}\times F_g\hat {r} =0$ el momento angular se conserva tal que, la fuerza gravitatorio es una fuerza central y por lo tanto no genera momento.

Energía Mecánica

b) Escriba una expresión para la energía mecánica identificando el potencial efectivo.

Primero voy a definir los datos que tengo disponibles


```python
#Datos
v_0= 2.0 #velocidad inicial
m=1.0 #masa de la partícula
r_0=0.1 #distancia de lanzamiento
M=5.0 # Masa del planeta
```

Ahora, daré una expresión general de Energía.

$E_m= \frac {1}{2}\cdot m \cdot \dot r ^2+\frac {1}{2}\cdot m \cdot r^2 \dot\theta ^2- \frac{G\cdot M \cdot m}{r} ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(1)$
                                       

Para la obtener el Potencial Efectivo, necesito que la expresión de energía quede expresada como una ecuación unidimensional. Para ello necesito reemplazar $\dot \theta$.

Conociendo la conservación del momento angular $\vec L^o$  (con origen "$o$" igual al centro del planeta), puedo llegar a una expresión para $\dot \theta$.

Por los datos iniciales, el momento angular inicial es

$ \vec L^o_i = r_0\hat r \times m\cdot v_0\hat\theta \Rightarrow  \vec L^o_i = r_0mv_0\hat z  ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(2)$

Ahora daré una expresión general del momento angular que cumplirá la conservación del momento angular
$ \vec L^o = r~ \hat r \times m\cdot (\dot r \hat r + r\dot\theta \hat \theta ) $

Por producto cruz, la velocidad radial se cancela con el radio sin generar momento.  
$ \Rightarrow \vec L^o = r~ \hat r \times m\cdot ( \not {\dot r \hat r} + r\dot\theta \hat \theta  ) \Rightarrow \vec L^o = r^2 m~ \dot \theta\hat z~~~~~~~~~~~~~~~~~~~~(3)$

Entonces por conservación del momento, podemos igualar las ecuaciones (2) y (3)

$r^2 m~ \dot \theta = r_0 mv_0 \Rightarrow \dot \theta= \frac {r_0 v_0}{r^2} ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(4)$

 Reemplazo (4) en la ecuación (1) de energía

  $E_m= \frac {1}{2}\cdot m \cdot \dot r ^2+\frac {1}{2}\cdot m \cdot r^{\not {2}} (\frac {r_0^2 v_0^2}{\not r^4})- \frac{G\cdot M \cdot m}{r}$

  $E_m= \frac {1}{2}\cdot m \cdot \dot r ^2+\underbrace{{\frac {mr_0^2 v_0^2}{ 2r^2}- \frac{G\cdot M \cdot m}{r}}}~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(5)\\ ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\color{lightgreen} {U_ef}$

Teniendo una ecuación unidimensional de enegía, puedo graficar el potencial efectivo


```python
import numpy as np
import matplotlib.pyplot as plt

 #Datos
v0 = 1 #velocidad inicial
m = 1 #masa de la partícula
r0 = 1 #distancia de lanzamiento
M = 6.7e10 # Masa del planeta
G = 6.7e-11

r = np.arange(0.01,10,0.01)
U = -(G* M* m)/r              # attractive energy term
T = (v0**2 * r0**2 *m)/(2* r**(2))    # repulsive energy term
Uef = U + T
plt.plot(r,U,r,T,r,Uef)
plt.xlim([ 0.01, 3])
plt.ylim([-15,15])
plt.title('Potencial efectivo')
plt.xlabel('r')
plt.ylabel('Uef')

plt.savefig('energy_vs_distance_curve.png', dpi=72)
plt.legend(['T','Uef','U'])
plt.grid(True)
plt.show()
```

c) Halle cuál tiene que ser el valor de $v_0$ para que la partrícula describa una  ́orbita circular.

Para que la partícula realize un movimiento circular con $v_0$ alrededor del planeta, la velocidad tangencial será nula, es decir $\dot r \hat r= 0$, por lo que la energía será puramente del potencial efectivo. Para llegar al movimiento circular, la energía debe ser el valor minimo de potencial, con la posición de máximo acercamiento $r_0$, que coincide con el extremo de la gráfica. Este lo podemos obtener derivando el potencial efectivo  $/ ~ \frac {∂U_{ef}}{∂r}|_{r_0} =0$

$\Rightarrow \frac {∂({\frac {mr_0^2 v_0^2}{ 2r^2} - \frac{G\cdot M \cdot m}{r}})}{∂r}= -{\frac {mr_0^2 v_0^2}{ r^3}+ \frac{G\cdot M \cdot m}{r^2}}|_{r_0} =0$

$-{\frac {mr_0^2 v_0^2}{ r_0^3}+ \frac{G\cdot M \cdot m}{r_0^2}} =0$

${\frac {\not m\not r_0^2 v_0^2}{\not r_0^3}= \frac{G\cdot M \cdot \not m}{\not r_0^2}}$

Despejando V, tenemos

$\color{orange}{ v_0 = \sqrt\frac{G\cdot M}{r_0}}$ velocidad $v_0$ para que el mocimiento de la partícula sea circular.
